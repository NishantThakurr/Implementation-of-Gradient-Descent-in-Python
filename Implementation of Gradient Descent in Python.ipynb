## Implementation of Gradient descent in python

```python
import numpy as np
```

```python
import pandas as pd
```

```python
##creating inputs and outputs
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
def gradient_descent(x, y, lr=0.01, ep=3000):
    m, b = 0.0, 0.0
    for epoch in range(ep):
        y_pred = m * x + b
        error = y - y_pred
        cost = np.mean(error ** 2)
        if cost == 0.0:
            print(f"Cost is 0.0 at epoch {epoch}. Stopping the gradient descent.")
            break
        dm = -2 * np.mean(error * x)
        db = -2 * np.mean(error)
        b -= db * lr
        m -= dm * lr
        print(f"Epoch {epoch}: Cost = {cost}, b = {b}, m = {m}")
```

```python
gradient_descent(x,y)
```
